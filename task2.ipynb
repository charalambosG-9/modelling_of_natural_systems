```python
#Code from the labs for 2D system

import numpy as np
import numpy.matlib 
import math
import matplotlib.pyplot as plt
import csv
from matplotlib import cm



class Numerical_methods:
    
    def __init__(self,f):  
        
        self.f=f
        self.x=[]
        self.t=[]
        self.N_dim=[]
        
        
    def Initialise(self,x_start,t_start):
        
        
        self.N_dim=np.shape(x_start)[0]
        self.x=x_start
        self.t=t_start
        
        
    def Euler(self,dt,N_iter):
        
        X=np.zeros([self.N_dim,N_iter])
        T=np.zeros([N_iter])
        
        X[:,0]=np.copy(self.x)            
        
        T[0]=np.copy(self.t)
        
        for n in range(1,N_iter):
        
            self.x=self.x+dt*f(self.x,self.t)
            X[:,n]=np.copy(self.x)
            
            self.t=self.t+dt
            T[n]=np.copy(self.t)
            
        return X, T
    
    def MidPoint(self,dt,N_iter):
        
        X=np.zeros([self.N_dim,N_iter])
        T=np.zeros([N_iter])
        
        X[:,0]=np.copy(self.x)
        T[0]=np.copy(self.t)
        
        for n in range(1,N_iter):
            
            x_mp=self.x+dt/2*f(self.x,self.t)
            self.x=self.x+dt*f(x_mp,self.t+dt/2)
            
            
            X[:,n]=np.copy(self.x)
            
            self.t=self.t+dt
            T[n]=np.copy(self.t)
            
        return X, T
    
    def RungeKutta2(self,dt,N_iter):
        
        X=np.zeros([self.N_dim,N_iter])
        T=np.zeros([N_iter])
        
        X[:,0]=np.copy(self.x)
        T[0]=np.copy(self.t)
        
        for n in range(1,N_iter):
            
            k1=f(self.x,self.t)
            k2=f(self.x+dt*k1,self.t+dt)    
                        
            self.x=self.x+dt*(k1+k2)/2
            
            X[:,n]=np.copy(self.x)
            
            self.t=self.t+dt
            T[n]=np.copy(self.t)
            
        return X, T

                
```

```python
def f(x,t):     #Here the function of interest needs to be set up. Equations for van der pol
                #oscilaltor were found and implemented.
               
    mu = 1  #This is a constant in the Van der Pol equation. Changing this value has a
            #large effect on sape of the output and is later varied to showcase this. 
    
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0] #these lines show the 2 defining equations of Van der Pol osscilator.
                                    #z[0] is equal to y which equals dx/dt and z[1] is dy/dt. x[0] is x and 
                                    #x[1] is y 
    
    return z


NM=Numerical_methods(f)           
dt=0.01                 #value of dt chosen so that the sytem converges.                   
N_iter=int(1000/dt)                
X_RK21=np.zeros([2,N_iter])
x_start=[1,1]     #starting vaules chosen that allow the periodic behaviour of the system to be shown.
t_start=0.
NM.Initialise(x_start,t_start)     
X_RK21,ts=NM.RungeKutta2(dt,N_iter) 




def f(x,t):     #Below code repeast the analysis but with differing values of the constant mu.
    
    mu = 10
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0]
    
    return z
NM=Numerical_methods(f)         
dt=0.01                          
N_iter=int(1000/dt)               
X_RK210=np.zeros([2,N_iter])
x_start=[1,1]     
t_start=0.
NM.Initialise(x_start,t_start)    
X_RK210,ts=NM.RungeKutta2(dt,N_iter) 




def f(x,t):     
    
    mu = 0.5
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0]
    
    return z
NM=Numerical_methods(f)         
dt=0.01                          
N_iter=int(1000/dt)               
X_RK205=np.zeros([2,N_iter])
x_start=[1,1]     
t_start=0.
NM.Initialise(x_start,t_start)    
X_RK205,ts=NM.RungeKutta2(dt,N_iter) 


def f(x,t):     
    
    mu = 5
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0]
    
    return z
NM=Numerical_methods(f)         
dt=0.01                          
N_iter=int(1000/dt)               
X_RK25=np.zeros([2,N_iter])
x_start=[1,1]     
t_start=0.
NM.Initialise(x_start,t_start)    
X_RK25,ts=NM.RungeKutta2(dt,N_iter) 


def f(x,t):     
    
    mu = 0
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0]
    
    return z
NM=Numerical_methods(f)         
dt=0.01                          
N_iter=int(1000/dt)               
X_RK20=np.zeros([2,N_iter])
x_start=[1,1]     
t_start=0.
NM.Initialise(x_start,t_start)    
X_RK20,ts=NM.RungeKutta2(dt,N_iter) 

def f(x,t):     
    
    mu = 2
    z=np.zeros([np.shape(x)[0]])
    
    z[0]=x[1]
    z[1]=mu*(1-x[0]**2)*x[1] - x[0]
    
    return z
NM=Numerical_methods(f)         
dt=0.01                          
N_iter=int(1000/dt)               
X_RK22=np.zeros([2,N_iter])
x_start=[1,1]     
t_start=0.
NM.Initialise(x_start,t_start)    
X_RK22,ts=NM.RungeKutta2(dt,N_iter) 



```
